Dot Product of 2 Vectors ::


```python
a=np.array([2,3])
b=np.array([4,5])
a_dot_b=np.dot(a,b)
a_cross_b=np.cross(a,b)
```

Projection of a vector over the b vector

OIP.jpg

```python
np.array([2,5])
b=np.array([8,-6])

#  magnitude of the 'v' vector :
magnitude_of_b=np.sqrt(sum(b**2))

proj_of_a_on_b=(np.dot(a,b)/magnitude_of_b**2)*b

print('Projection of a Vector on v vector=',proj_of_a_on_b)

```

```python
#  PROJECTION OF V1 ON V2 ::

v1=np.array([50,60,100])
v2=np.array([1,2,3])

#  magnitude of v2 vector :
magnitude_of_v2=np.sqrt(sum(v2**2))

proj_of_v1_on_v2=(np.dot(v1,v2)/magnitude_of_v2**2)*v2

print('Projection of v1 on v2 is:',proj_of_v1_on_v2)
```
